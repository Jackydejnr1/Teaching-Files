```python
import random

print("ðŸŽ² Welcome to the Math Guessing Game ðŸŽ²")
print("You will be asked some math questions. You have 3 attempts per question.\n")

# Number of questions
num_questions = random.randint(5, 10)  # random number of questions between 5 and 10
score = 0
questions = []  # store questions and answers

for i in range(1, num_questions + 1):
    # Random numbers and operator
    a = random.randint(1, 12)
    b = random.randint(1, 12)
    operator = random.choice(["+", "-", "*", "/"])

    # Ensure no decimals for division
    if operator == "/":
        a = a * b  # so a is divisible by b

    # Correct answer
    if operator == "+":
        answer = a + b
    elif operator == "-":
        answer = a - b
    elif operator == "*":
        answer = a * b
    elif operator == "/":
        answer = a // b

    # Store question for review later
    questions.append((f"{a} {operator} {b}", answer))

    # Ask user with up to 3 attempts
    correct = False
    for attempt in range(1, 4):  # 3 tries
        try:
            user_answer = int(input(f"Q{i}: What is {a} {operator} {b}? (Attempt {attempt}/3): "))
        except ValueError:
            print("Please enter a valid number.")
            continue

        if user_answer == answer:
            print("âœ… Correct!\n")
            score += 1
            correct = True
            break
        else:
            if attempt < 3:
                print("âŒ Wrong, try again.")
            else:
                print("âŒ Wrong. Moving to the next question.\n")

# Game Over Summary
print("\nðŸ“Š Game Over! Here are the results:")

# Show correct answers
for idx, (q, ans) in enumerate(questions, start=1):
    print(f"Q{idx}: {q} = {ans}")

# Show final score & percentage
percentage = (score / num_questions) * 100
print(f"\nâœ… You got {score} out of {num_questions} correct.")
print(f"ðŸ“ˆ Your score: {percentage:.2f}%")
```

```python
import random

print("\nðŸŽ® Welcome to Rock-Paper-Scissors ðŸŽ®")
print("We will play 10 rounds. Try to beat the computer!\n")

choices = ["rock", "paper", "scissors"]
player_score = 0
computer_score = 0

for round_no in range(1, 11):
    print(f"\nRound {round_no}")
    player = input("Enter rock, paper, or scissors: ").lower()

    if player not in choices:
        print("Invalid choice! You lose this round.")
        computer_score += 1
        continue

    computer = random.choice(choices)
    print("Computer chose:", computer)

    # Decide winner
    if player == computer:
        print("It's a draw!")
    elif (player == "rock" and computer == "scissors") or \
         (player == "scissors" and computer == "paper") or \
         (player == "paper" and computer == "rock"):
        print("You win this round! ðŸŽ‰")
        player_score += 1
    else:
        print("Computer wins this round! ðŸ¤–")
        computer_score += 1

# Final result
print("\nðŸ† Final Results ðŸ†")
print(f"Your Score: {player_score}")
print(f"Computer Score: {computer_score}")

if player_score > computer_score:
    print("ðŸŽ‰ Congratulations, you win the game!")
elif player_score < computer_score:
    print("ðŸ¤– Computer wins the game. Better luck next time!")
else:
    print("ðŸ˜… It's a tie!")
```

```python
print("ðŸ›ï¸ Welcome to the Sales Tax & Import Duty Calculator ðŸ›ï¸\n")

items = []
sales_taxes = 0
total = 0

while True:
    qty = int(input("Quantity: "))
    name = input("Product Name: ")
    price = float(input("Price: "))
    imported = input("Is it Imported [y/n]: ").lower()
    exempted = input("Is it Exempted [y/n]: ").lower()

    # Calculate tax
    tax = 0
    if exempted != "y":   # basic sales tax (10%)
        tax += 0.10 * price
    if imported == "y":   # import duty (5%)
        tax += 0.05 * price

    item_tax = round(tax, 2) * qty
    final_price = (price + round(tax, 2)) * qty

    items.append((qty, name, final_price))

    sales_taxes += item_tax
    total += final_price

    more = input("Add More Products [y/n]: ").lower()
    if more != "y":
        break

# Print receipt
print("\nðŸ§¾ Receipt:")
for qty, name, price_with_tax in items:
    print(f"{qty} {name} at {price_with_tax:.2f}")

print(f"\nSales Taxes: {sales_taxes:.2f}")
print(f"Total: {total:.2f}")
```
